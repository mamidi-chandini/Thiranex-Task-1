```python
import pandas as pd

data = {
    "Date":["2026-01-01","2026-01-02","2026-01-03","2026-01-04","2026-01-05",
            "2026-01-06","2026-01-07","2026-01-08","2026-01-09","2026-01-10",
            "2026-01-11","2026-01-12","2026-01-13","2026-01-14","2026-01-15",
            "2026-01-16","2026-01-17","2026-01-18","2026-01-19","2026-01-20"],
    "Product":["Laptop","Phone","Shirt","Rice","Shoes","TV","Milk","Watch","Bag","Jeans",
               "Headphones","Sugar","Sandals","Tablet","Jacket","Bread","Cap","Socks","Monitor","Keyboard"],
    "Category":["Electronics","Electronics","Clothing","Grocery","Footwear","Electronics","Grocery",
                "Accessories","Accessories","Clothing","Electronics","Grocery","Footwear",
                "Electronics","Clothing","Grocery","Accessories","Clothing","Electronics","Electronics"],
    "Quantity":[1,2,3,5,2,1,10,2,1,2,3,4,2,1,1,8,3,5,2,4],
    "Price":[50000,20000,800,60,2500,35000,30,3000,1500,1200,1500,45,1000,25000,2500,40,500,200,12000,1000],
    "Customer_Age":[25,30,22,45,28,40,35,27,24,29,26,38,31,33,36,21,23,20,34,32],
    "Payment_Method":["UPI","Card","Cash","UPI","Card","UPI","Cash","Card","UPI","Cash",
                      "Card","UPI","Cash","UPI","Card","Cash","UPI","Card","UPI","Cash"]
}

df = pd.DataFrame(data)
df["Total_Sales"] = df["Quantity"] * df["Price"]

df.head()
```

```python
# Display first 5 rows
print(df.head())

# Dataset information
print("\nDataset Information:")
print(df.info())

# Statistical summary
print("\nSummary Statistics:")
print(df.describe())

# Total Sales
print("\nTotal Sales:", df["Total_Sales"].sum())

# Category-wise Sales
print("\nCategory-wise Sales:")
print(df.groupby("Category")["Total_Sales"].sum())

# Payment Method Count
print("\nPayment Method Count:")
print(df["Payment_Method"].value_counts())
```

```python
import matplotlib.pyplot as plt

# 1. Category-wise Sales (Bar Chart)
category_sales = df.groupby("Category")["Total_Sales"].sum()

plt.figure(figsize=(8,5))
category_sales.plot(kind="bar")
plt.title("Category-wise Sales")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()

# 2. Payment Method Distribution (Pie Chart)
payment_counts = df["Payment_Method"].value_counts()

plt.figure(figsize=(6,6))
payment_counts.plot(kind="pie", autopct="%1.1f%%")
plt.title("Payment Method Distribution")
plt.ylabel("")
plt.show()

# 3. Customer Age Distribution (Histogram)
plt.figure(figsize=(8,5))
plt.hist(df["Customer_Age"], bins=6)
plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.show()

# 4. Top Products by Sales
product_sales = df.groupby("Product")["Total_Sales"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
product_sales.plot(kind="bar")
plt.title("Top Products by Sales")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```
