```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df = df.drop_duplicates()
print("Duplicates removed successfully!")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.hist(df['total_bill'])
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Count")
plt.show()
```

```python
# Relationship between Total Bill and Tip

import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.scatter(df['total_bill'], df['tip'])

plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Relationship between Total Bill and Tip")

plt.show()
```

```python
# Basic statistics of Total Bill

df['total_bill'].describe()
```

```python
# Distribution of total bill using histogram

import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df['total_bill'], bins=20)
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.title("Distribution of Total Bill")
plt.show()
```

```python
# Relationship between Total Bill and Tip

plt.figure(figsize=(8,5))
plt.scatter(df['total_bill'], df['tip'])
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Total Bill vs Tip")
plt.show()
```

```python
# Total bill distribution by day

plt.figure(figsize=(8,5))
df.groupby('day')['total_bill'].mean().plot(kind='bar')
plt.xlabel("Day")
plt.ylabel("Average Total Bill")
plt.title("Average Total Bill by Day")
plt.show()
```

```python
# Average total bill by gender

plt.figure(figsize=(8,5))
df.groupby('sex')['total_bill'].mean().plot(kind='bar')
plt.xlabel("Gender")
plt.ylabel("Average Total Bill")
plt.title("Average Total Bill by Gender")
plt.show()
```

```python
# Average total bill by smoker status

plt.figure(figsize=(8,5))
df.groupby('smoker')['total_bill'].mean().plot(kind='bar')
plt.xlabel("Smoker Status")
plt.ylabel("Average Total Bill")
plt.title("Average Total Bill by Smoker Status")
plt.show()
```

# Task 1 Observations

1. Total Bill distribution shows that most customers spend a moderate amount, while few customers have very high bills.

2. The scatter plot between Total Bill and Tip shows a positive relationship. As the total bill increases, the tip amount generally increases.

3. Day wise analysis shows the average spending pattern of customers on different days.

4. Gender wise analysis helps compare the average bill amount between male and female customers.

5. Smoker and non-smoker analysis helps understand spending behavior based on smoking status.

```python
# Dashboard of Key Findings

fig, axes = plt.subplots(2, 2, figsize=(12,8))

# Total Bill Distribution
axes[0,0].hist(df['total_bill'], bins=20)
axes[0,0].set_title("Total Bill Distribution")
axes[0,0].set_xlabel("Total Bill")
axes[0,0].set_ylabel("Frequency")

# Total Bill vs Tip
axes[0,1].scatter(df['total_bill'], df['tip'])
axes[0,1].set_title("Total Bill vs Tip")
axes[0,1].set_xlabel("Total Bill")
axes[0,1].set_ylabel("Tip")

# Average Bill by Day
df.groupby('day')['total_bill'].mean().plot(kind='bar', ax=axes[1,0])
axes[1,0].set_title("Average Bill by Day")
axes[1,0].set_xlabel("Day")
axes[1,0].set_ylabel("Average Total Bill")

# Average Bill by Gender
df.groupby('sex')['total_bill'].mean().plot(kind='bar', ax=axes[1,1])
axes[1,1].set_title("Average Bill by Gender")
axes[1,1].set_xlabel("Gender")
axes[1,1].set_ylabel("Average Total Bill")

plt.tight_layout()
plt.show()
```
